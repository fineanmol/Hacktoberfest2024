## How do I filter rows of a pandas DataFrame by column value?

```python
booleans = []
for length in movies.duration:
    if length >= 200:
        booleans.append(True)
    else :
        booleans.append(False)
```

```python
booleans[0:5]
```

```python
is_long = pd.Series(booleans)
```

```python
is_long.head()
```

```python
movies[is_long]
```

```python
# Simpler code direct
is_long_short = movies.duration >= 200
is_long_short.head()
```

```python
#Even simpler
movies[movies.duration >= 200]
```

## How do i apply multiple filters to DataFrame?

```python
movies[(movies.duration >= 200) & (movies.genre == 'Drama')]
#You have to use parentheses and ampersand/pipe
```

```python
movies[movies.genre.isin(['Crime', 'Drama', 'Action'])]
```
