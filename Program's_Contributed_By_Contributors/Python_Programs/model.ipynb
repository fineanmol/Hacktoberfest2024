```python
# importing dependiences 
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

print("import successful")
```

```python
# importing data to pandas dataframe
sonar_data = pd.read_csv(r"C:\Users\mujee\Machine Learning\rock_or_mine\data.csv",header=None)
print("reading data successfull")
```

```python
sonar_data.groupby(60).mean()
```

```python
# splitting features and output lables
x = sonar_data.drop(columns=60,axis=1)
y = sonar_data[60]
```

```python
print(x)
print(y)
```

```python
# splittng training and testing data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,stratify=y,random_state=1)
print(x_train.shape)
```

```python
# model creation
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
# check accuracy of model by checking it on train data
x_train_prediction = model.predict(x_train)
x_train_prediction_accuracy = accuracy_score(x_train_prediction,y_train)
print("accuracy is " , x_train_prediction_accuracy)
```

```python
# teting the accuracy of the model
x_test_prediction = model.predict(x_test)
x_test_prediction_accuracy = accuracy_score(x_test_prediction,y_test)
print("accuracy is " , x_test_prediction_accuracy)
```
