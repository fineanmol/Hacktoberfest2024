```python
t=int(input())
for i in range(0,t):
    m=int(input())
    if(m>=3600000):
        h=int(m/3600000)
        m=m-h*3600000
        print("{}hrs".format(h),end=" ")
    if(m>=60000):
        v=int(m/60000)
        m=m-v*60000
        print("{}min".format(v),end=" ")
    if(m>=1000):
        l=int(m/1000)
        m=m-l*1000
        print("{}sec".format(l),end=" ")
    print("{}milis".format(m),end="\n")
```

```python
vowels = 'aeiou'
consonants='bcdfghjklmnpqrstvwxyz'
ls=["Hello",
"Test"
]
for i in ls:
    ip_str = i
    ip_str = ip_str.casefold()
    count = {}.fromkeys(vowels,0)
    ss = {}.fromkeys(consonants,0)
    v=0
    c=0
    for char in ip_str:
        if char in count:
            v += 1
        elif char in ss:
            c += 1
    print("V: "+str(v)+" C: "+str(c))
#print("C : "+str(c))
```

```python
ls=[2,4,6,12]
for i in ls:
    n = i

    # upward pyramid
    for i in range(n):
        for j in range(n - i - 1):
            print(' ', end='')
        for j in range(2 * i + 1):
            print(chr(65 + j), end='')
        print()

    # downward pyramid
    for i in range(n - 1):
        for j in range(i + 1):
            print(' ', end='')
        for j in range(2*(n - i - 1) - 1):
            print(chr(65 + j), end='')
        print()
    print()
    print()
```

```python
row = 12
# Upper part of hollow diamond
for i in range(1, row+1):
    for j in range(1,row-i+1):
        print(" ", end="")
    for j in range(1, 2*i):
        if j==1 or j==2*i-1:
            print("*", end="")
        else:
            print(" ", end="")
    print()

# Lower part of hollow diamond
for i in range(row-1,0, -1):
    for j in range(1,row-i+1):
        print(" ", end="")
    for j in range(1, 2*i):
        if j==1 or j==2*i-1:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```

```python
import random

def bubbleSort(arr,wt,rt):
    n = len(arr)
    for i in range(n-1):
        for j in range(0, n-i-1):
            if arr[j] > arr[j + 1] :
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                wt[j], wt[j + 1] = wt[j + 1], wt[j]
                rt[j], rt[j + 1] = rt[j + 1], rt[j]



ls=[2,3,4,6,15]
for i in ls:
    pr=[]
    wt=[]
    print(i)
    for k in range(0,i):
        ls=[]
        we=[]
        n=random.randrange(3, 23)
        for j in range(0,n):
            we.append(int(random.randrange(0, 35)))
            ls.append(int(random.randrange(0, 35)))
        pr.append(ls)
        wt.append(we)
    rt=[]
    for k in range(0,len(wt)):
        ls=[]
        for j in range(0,len(wt[k])):
            ls.append(pr[k][j]/wt[k][j])
        rt.append(ls)
        
    for k in range(0,len(wt)):
        bubbleSort(rt[0],wt[0],pr[0])
    
    print(rt[0])
    print(pr[0])
    print(ls[0])
```
